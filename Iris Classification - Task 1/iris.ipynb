The Iris flower data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

This dataset became a typical test case for many statistical classification techniques in machine learning such as support vector machines.

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">Percentile Method - Removing Ouliers</p>

```python
# Function to detect outliers using Percentiles
def detect_outliers(df, features):
    outliers = []
    for feature in features:
        lower_bound = np.percentile(df[feature], 1)
        upper_bound = np.percentile(df[feature], 99)
        outliers.extend(df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index)
    return list(set(outliers))

# Detect and remove outliers
outliers = detect_outliers(df, features)
df = df.drop(outliers).reset_index(drop=True)

# Print the result
print("Outliers removed. New data shape:", df.shape)
```

```python
# histograms of all numeric variables with hue as Class
plt.figure(figsize=(20, 15))
for i, column in enumerate(df[features], 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df, x=column, kde=True, palette='coolwarm', element="step")
    plt.title(f'Histogram of {column}')
plt.tight_layout()
plt.show()

```

```python
# separate features and target
X = df.drop('species', axis=1)
y = df['species']

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=29)

# scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Training data shape:", X_train_scaled.shape)
print("Testing data shape:", X_test_scaled.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">Random Forest</p>

```python
# Random Forest with Grid Search
rfc_param_grid = {
    'n_estimators': [100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [4, 5, 6, 7, 8],
    'criterion': ['gini', 'entropy']
}

rfc = RandomForestClassifier()
rfc_cv = GridSearchCV(rfc, rfc_param_grid, cv=3, scoring='accuracy')
rfc_cv.fit(X_train_scaled, y_train)

rfc_prediction = rfc_cv.predict(X_test_scaled)

print("Tuned Hyperparameters (Best Parameters): ", rfc_cv.best_params_)
print("Accuracy: ", rfc_cv.best_score_)

cr_rfc = classification_report(y_test, rfc_prediction)
print(cr_rfc)
```

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">Support Vector Machine</p>

```python
# Support Vector Machine with Grid Search
svm_param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

svm = SVC()
svm_cv = GridSearchCV(svm, svm_param_grid, cv=10, scoring='accuracy')
svm_cv.fit(X_train_scaled, y_train)

svm_prediction = svm_cv.predict(X_test_scaled)

print("Tuned Hyperparameters (Best Parameters): ", svm_cv.best_params_)
print("Accuracy: ", svm_cv.best_score_)

cr_svm = classification_report(y_test, svm_prediction)
print(cr_svm)

```

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">Decision Tree</p>

```python
# Decision Tree with Grid Search
tree_param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [4, 5, 6, 7, 8],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

tree = DecisionTreeClassifier()
tree_cv = GridSearchCV(tree, tree_param_grid, cv=10, scoring='accuracy')
tree_cv.fit(X_train_scaled, y_train)

tree_prediction = tree_cv.predict(X_test_scaled)

print("Tuned Hyperparameters (Best Parameters): ", tree_cv.best_params_)
print("Accuracy: ", tree_cv.best_score_)

cr_tree = classification_report(y_test, tree_prediction)
print(cr_tree)

```

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">K-Nearest Neighbor</p>

```python
# K-Nearest Neighbors with Grid Search
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, knn_param_grid, cv=10, scoring='accuracy')
knn_cv.fit(X_train_scaled, y_train)

knn_prediction = knn_cv.predict(X_test_scaled)

print("Tuned Hyperparameters (Best Parameters): ", knn_cv.best_params_)
print("Accuracy: ", knn_cv.best_score_)

cr_knn = classification_report(y_test, knn_prediction)
print(cr_knn)
```

### <p style="font-family: 'italic', cursive; font-size: 24px; font-weight: bold; text-align: center; color: skyblue; background-color: #212529; padding: 10px 20px 10px 10px; border: 2px solid #343a40; border-radius: 10px; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);">Model Selection</p>

```python
# Collecting model performance metrics
models = ['Random Forest', 'Support Vector Machine', 'K-Nearest Neighbors', 'Decision Tree']
accuracy_scores = [
    accuracy_score(y_test, rfc_prediction),
    accuracy_score(y_test, svm_prediction),
    accuracy_score(y_test, knn_prediction),
    accuracy_score(y_test, tree_prediction)
]

precision_scores = [
    precision_score(y_test, rfc_prediction, average='macro'),
    precision_score(y_test, svm_prediction, average='macro'),
    precision_score(y_test, knn_prediction, average='macro'),
    precision_score(y_test, tree_prediction, average='macro')
]

recall_scores = [
    recall_score(y_test, rfc_prediction, average='macro'),
    recall_score(y_test, svm_prediction, average='macro'),
    recall_score(y_test, knn_prediction, average='macro'),
    recall_score(y_test, tree_prediction, average='macro')
]

f1_scores = [
    f1_score(y_test, rfc_prediction, average='macro'),
    f1_score(y_test, svm_prediction, average='macro'),
    f1_score(y_test, knn_prediction, average='macro'),
    f1_score(y_test, tree_prediction, average='macro')
]

# Creating a DataFrame to display the results
scores_df = pd.DataFrame({
    'Accuracy': accuracy_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1 Score': f1_scores
}, index=models)

print(scores_df)

# Plotting the model performance comparison
scores_df.plot(kind='bar', figsize=(14, 8), colormap='coolwarm')
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.show()
```

```python
# Selecting the best model based on accuracy
model_scores = {
    'Random Forest': rfc_cv.best_score_,
    'Support Vector Machine': svm_cv.best_score_,
    'K-Nearest Neighbors': knn_cv.best_score_,
    'Decision Tree': tree_cv.best_score_
}

best_model_name = max(model_scores, key=model_scores.get)
best_model_score = model_scores[best_model_name]

print(f"Best model is {best_model_name} with a score of {best_model_score}")

if best_model_name == 'Random Forest':
    best_params = rfc_cv.best_params_
elif best_model_name == 'Support Vector Machine':
    best_params = svm_cv.best_params_
elif best_model_name == 'K-Nearest Neighbors':
    best_params = knn_cv.best_params_
else: 
    best_params = tree_cv.best_params_

print(f"Best parameters for {best_model_name}: {best_params}")
```
